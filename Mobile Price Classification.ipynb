##**SPLITING DATA INTO TRAINING AND TESTING SETS**

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.metrics import precision_score, recall_score, f1_score
from sklearn.metrics import classification_report
```

##**CREATING LOGISTIC REGRESSION MODEL**

```python
from sklearn.linear_model import LogisticRegression
classifier=LogisticRegression()
```

##**TRAINING LOGISTIC REGRESSION MODEL**

```python
classifier.fit(x_train,y_train)
```

##**MAKING PREDICTION LOGISTIC REGRESSION MODEL**

```python
y_pred=classifier.predict(x_test)
```

##**APPLYING CONFUSION MATRIX MODEL FITTING**

```python
from sklearn.metrics import confusion_matrix
cm= confusion_matrix(y_test,y_pred)
cm
```

##**CLASSIFICATION MODEL ACCURACY SCORE**

```python
from sklearn.metrics import accuracy_score
accuracy=accuracy_score(y_test,y_pred)
accuracy
```

##**CLASSIFICATION REPORT OF LOGISTIC REGRESSION**

```python
print(classification_report(y_test,y_pred))
```

##**CREATING KNEIGHBOURS CLASSIFIER (KNN)**

```python
from sklearn.neighbors import KNeighborsClassifier
classifier= KNeighborsClassifier(n_neighbors=5, metric='minkowski', p=2 )
```

##**TRAINING KNEIGHBOURS CLASSIFIER (KNN)**

```python
classifier.fit(x_train, y_train)
```

##**MAKING PREDICTIONS IN KNN**

```python
y_pred= classifier.predict(x_test)
```

##**APPLYING CONFUSION MATRIX IN KNN**

```python
from sklearn.metrics import confusion_matrix
cm= confusion_matrix(y_test, y_pred)
cm
```

## **CLASSIFYING KNN ACCURACY SCORE**

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

##**CLASSIFICATION REPORT OF KNEAREST NEIGHBORS**

```python
print(classification_report(y_test,y_pred))
```

##**CREATING SUPPORT VECTOR CLASSIFIER (SVM)**

```python
from sklearn.svm import SVC # "Support vector classifier"
classifier = SVC(kernel='linear', random_state=0)
```

##**TRAINING SUPPORT VECTOR CLASSIFIER (SVM)**

##**MAKING PREDICTIONS OF SVM**

##**APPLYING CONFUSION MATRICS IN SVM**

```python
from sklearn.metrics import confusion_matrix
cm= confusion_matrix(y_test, y_pred)
cm
```

##**CLASSIFYING SVM ACCURACY SCORE**

##**COMPARISON BETWEEN LOGISTIC REGRESSION,KNN AND SVM**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample accuracy data
accuracy_data = {
    'Model': ['Logistic regression', 'KNN','SVM'],
    'Accuracy': [0.62, 0.95 , 0.97,]  # Sample accuracy scores
}

# Create DataFrame
accuracy_df = pd.DataFrame(accuracy_data)

# Define colors for each bar
colors = ['blue', 'green', 'red',]

# Plotting
plt.figure(figsize=(10, 6))
bars = plt.bar(accuracy_df['Model'], accuracy_df['Accuracy'], color=colors)
plt.title('Model Accuracy Comparison')
plt.xlabel('Model')
plt.ylabel('Accuracy')
plt.ylim(0, 1)  # Set y-axis limits between 0 and 1
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent clipping of labels

# Add values on top of the bars
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        height,
        f'{height*100:.0f}%',  # Convert to percentage and format
        ha='center',
        va='bottom'
    )

plt.show()
```
